```python
import matplotlib.pyplot as plt
```

```python
def maximal_points_left_to_right(points):
    # Initialize the maximal points list with the first point
    maximal_points = [points[0]]
    comparisons = 0

    # Iterate through the points from the second one
    for i in range(1, len(points)):
        comparisons += 1
        # Compare the y-coordinate of the current point with the last maximal point
        if points[i][1] > maximal_points[-1][1]:
            # If the current point has a greater y-coordinate, add it to the maximal points list
            maximal_points.append(points[i])

    return maximal_points, comparisons
```

```python
def maximal_points_right_to_left(points):
    # Initialize the maximal points list with the last point
    maximal_points = [points[-1]]
    comparisons = 0

    # Iterate through the points from the second to the last one in reverse order
    for i in range(len(points) - 2, -1, -1):
        comparisons += 1
        # Compare the y-coordinate of the current point with the last maximal point
        if points[i][1] > maximal_points[-1][1]:
            # If the current point has a greater y-coordinate, add it to the maximal points list
            maximal_points.append(points[i])

    # Reverse the list to get the correct order of maximal points
    return maximal_points[::-1], comparisons
```

```python
def visualize_points(points, maximal_points, comparisons, variant):
    # Extract x and y coordinates for input and maximal points
    x, y = zip(*points)
    mx, my = zip(*maximal_points)

    # Create a figure with a size of 10x6 inches
    plt.figure(figsize=(10, 6))

    # Plot input points
    plt.scatter(x, y, color='blue', s=80, label='Input Points')

    # Highlight maximal points
    plt.scatter(mx, my, color='red', s=120, label='Maximal Points')

    # Draw dashed lines connecting the points in the Maximal Points process
    for i in range(len(maximal_points) - 1):
        plt.plot([maximal_points[i][0], maximal_points[i+1][0]],
                 [maximal_points[i][1], maximal_points[i+1][1]],
                 linestyle='--', color='gray', alpha=0.7)

    # Annotate comparisons at each step
    for i, point in enumerate(maximal_points):
        plt.annotate(f'Comp: {i + 1}', (point[0], point[1] + 0.5), color='black', fontsize=8)

    # Set plot title and axis labels
    plt.title(f'Maximal Points Visualization ({variant} Variant)')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # Display legend, grid, and the plot
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

```python
def main():
    # Get the number of points from the user
    n = int(input("Enter the number of points: "))
    points = []

    # Input the coordinates for each point
    for i in range(n):
        x, y = map(int, input(f"Enter (x, y) for point {i + 1}: ").split())
        points.append((x, y))

    # Apply the left-to-right and right-to-left variants of the Maximal Points algorithm
    left_to_right_maximal_points, left_to_right_comparisons = maximal_points_left_to_right(points)
    right_to_left_maximal_points, right_to_left_comparisons = maximal_points_right_to_left(points)

    # Display input points and results
    print("\nInput Points:")
    for i, point in enumerate(points):
        print(f"Point {i + 1}: {point}")

    print("\nMaximal Points (Left to Right):")
    for i, point in enumerate(left_to_right_maximal_points):
        print(f"Point {i + 1}: {point}")
    print(f"Number of comparisons (Left to Right): {left_to_right_comparisons}")

    print("\nMaximal Points (Right to Left):")
    for i, point in enumerate(right_to_left_maximal_points):
        print(f"Point {i + 1}: {point}")
    print(f"Number of comparisons (Right to Left): {right_to_left_comparisons}")

    # Visualize the input and maximal points for both variants
    visualize_points(points, left_to_right_maximal_points, left_to_right_comparisons, 'Left-to-Right')
    visualize_points(points, right_to_left_maximal_points, right_to_left_comparisons, 'Right-to-Left')
```

```python
if __name__ == "__main__":
    main()
```
